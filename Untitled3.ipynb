```python

# Import required libraries
import requests
import pandas as pd

# Raw GitHub JSON URL
url = "https://raw.githubusercontent.com/marwokaclinton-ops/kq-employee-dataset/refs/heads/main/kq%20employee%20.json"

# Send request to GitHub
response = requests.get(url)

# Convert response to JSON
data = response.json()

# Display extracted JSON data
print(data)
```

```python
# Convert JSON data into DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print(df.head())
```

```python
# Save DataFrame as CSV
df.to_csv("employees.csv", index=False)

print("CSV file saved successfully.")
```

```python
print("JSON data extracted and saved as CSV successfully.")
```
